What is Correlation (in Pandas)?

Correlation tells you how strongly two variables are related.

In Pandas, you usually compute it using:

df.corr()

This gives a correlation matrix between all numeric columns.

Intuition (very important)

Think like this:
| Scenario                              | Correlation       |
| ------------------------------------- | ----------------- |
| As one increases, the other increases | **Positive (+1)** |
| As one increases, the other decreases | **Negative (-1)** |
| No relationship                       | **0**             |


```python
'''example for positive correlation'''
import pandas as pd

data = {
    "hours_studied": [1, 2, 3, 4, 5],
    "marks": [30, 40, 50, 60, 70]
}

df = pd.DataFrame(data)

print(df.corr())
```

```python
'''example for negative correlation'''
data = {
    "hours_studied": [1, 2, 3, 4, 5],
    "marks": [90, 80, 70, 60, 50]
}
df=pd.DataFrame(data)
print(df.corr())
```

Why is Correlation Important?

1. Feature Selection (VERY IMPORTANT in ML)
If you're building an ML model:
You want variables that actually influence your target
Correlation helps you find that

Example:
df.corr()["price"]

👉 You can see which features affect price


2. Remove Redundant Data
If two features are highly correlated:

height ↔ weight (very similar info)

👉 Keeping both can:
confuse models
cause multicollinearity

So you remove one.

3. Understand Relationships in Data
Before modeling, you should always ask:

What affects what?
Which variables move together?

Correlation gives quick answers.

4. Detect Patterns (EDA - Exploratory Data Analysis)
Correlation is one of the first steps in data analysis.
👉 You get a visual map of relationships.

```python
# let's consider the health data and find correlation between numeric columns
import pandas as pd
df=pd.read_csv('data_dataframe/health_data.csv')
print(df.corr())
# Note: The corr() method ignores "not numeric" columns.
```

The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.
